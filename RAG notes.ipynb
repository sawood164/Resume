## RAG (Retrieval-Augmented Generation) Systems

- RAG is like an AI that "retrieves before it answers".
- A well-designed flow makes RAG systems easy to build and maintain.
- It pulls facts from a knowledge source (not just guesses).
- Can respond in multiple formats — text, audio, etc.
- Enterprise RAG systems give more context-aware, accurate replies.
- Web-based RAG setups sometimes struggle with compute limits.
- More context = smarter responses.
- Powered by Seq2Seq models and Contrastive Loss.

## AI Architecture & Optimization

- Memory systems (like RAM boards) are at the core.
- Vectors drive most modern AI processes.
- Bad infra = expensive mistakes.
- Need to scale smartly, even with limited hardware.
- Always-On systems require efficient setups.
- Cache is critical for performance speed.
- About 10% system duplication happens due to rigid schemas.
- Optimization helps reduce power use and improves system flow.

## Reasoning & Attention

- We don’t fully understand how AI “reasons” yet.
- DeepMind’s Chain-of-Thought models were key breakthroughs.
- Attention Mechanism helps AI focus on the right data.
- Introduced in the 2017 paper: “Attention is All You Need”.
- Attention = better accuracy, especially in multi-agent systems.
- Stronger attention + cleaner data = smarter AI.
