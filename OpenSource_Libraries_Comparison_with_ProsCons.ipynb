## üîç Category-wise Summary

| **Data Access**       | `datasets`                                          | Easy access to shared ML datasets, mostly NLP-related |


## üß† Pros & Cons of Key Open-Source Libraries 

| Library / Framework       | ‚úÖ Pros                                                                 | ‚ùå Cons                                                                |
|---------------------------|-------------------------------------------------------------------------|------------------------------------------------------------------------|
| `ruamel.yaml`             | Preserves comments, formatting, and order in YAML files                 | Heavier and more complex than PyYAML                                  |
| `rich`                    | Beautiful terminal formatting and tracebacks                            | Adds overhead if only minimal output is needed                        |
| `decorator`               | Simplifies decorator syntax                                              | Minimal niche; may not be necessary for all use cases                 |
| `numpy`                   | Fast array operations, supports broadcasting                            | Limited for tabular/mixed data structures                             |
| `pandas`                  | Excellent for tabular and time series data                              | Slower for very large datasets; high memory usage                     |
| `matplotlib`              | Highly customizable plotting                                             | Verbose and less intuitive than newer libraries                       |
| `seaborn`                 | Beautiful statistical plots with minimal code                           | Limited interactivity                                                  |
| `scipy`                   | Wide range of scientific computing tools                                | API can feel fragmented; requires `numpy` knowledge                   |
| `torch` (PyTorch)         | Dynamic computation graph, GPU support                                  | Steeper learning curve than `sklearn` for beginners                   |
| `datasets` (HF)           | Access to preprocessed ML datasets                                      | Mostly NLP focused; less for tabular/vision datasets                  |
| `sentence_transformers`   | Easy sentence-level embeddings                                           | Heavy models; slow without GPU                                        |
| `sklearn`                 | Simple API for ML models and pipelines                                  | Not suitable for deep learning tasks                                  |
| `umap`                    | Fast, effective dimensionality reduction                                | Results can vary due to randomness; not easily interpretable          |
| `dotenv`                  | Secure environment variable management                                  | Requires `.env` discipline; no validation by default                  |
| `plotly`                  | Interactive visualizations and dashboards                               | Can be slow in Jupyter with many plots                                |
| `altair`                  | Declarative syntax, fast prototyping                                    | Limited support for very large datasets                               |
| `IPython`                 | Rich dev experience and notebook support                                | Adds bulk to small scripts; overlaps with standard Python shell       |
